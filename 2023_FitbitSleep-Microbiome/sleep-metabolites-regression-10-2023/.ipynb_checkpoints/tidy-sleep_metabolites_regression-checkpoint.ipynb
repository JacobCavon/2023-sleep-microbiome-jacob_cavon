```python
import pandas as pd
import json
from arivale_data_interface import get_snapshot
import warnings
warnings.simplefilter("ignore")

#import the df and the feature indices

sleep_merged = pd.read_csv('../../sleep_microbe_metabolite_df/sleep_microbes_metabolites_df.csv')

with open('../../sleep_microbe_metabolite_df/activity_features.json', 'r') as f:
    activity_features = json.load(f)
with open('../../sleep_microbe_metabolite_df/metabolite_features.json', 'r') as f:
    metabolite_features = json.load(f)
with open('../../sleep_microbe_metabolite_df/sleep_features.json', 'r') as f:
    sleep_features = json.load(f)
with open('../../sleep_microbe_metabolite_df/microbe_features.json', 'r') as f:
    microbe_features = json.load(f)
```

```python
import numpy as np

#log transform and rename the metabolite columns
sleep_merged[metabolite_features] = np.log(sleep_merged[metabolite_features])

#log transform the sleep columns
sleep_merged[sleep_features] = np.log(sleep_merged[sleep_features])

#log transform the activity columns
sleep_merged[activity_features] = np.log(sleep_merged[activity_features])
```

```python
sleep_merged[activity_features].hist()
```

# Regression Metabolites w/ sex, age, and BMI covariates 

```python
from rich.progress import track
from statsmodels.formula.api import ols
from statsmodels.stats.multitest import multipletests

def get_results(dependent_feature, independent_feature):
    """Get a single association for a specific combination of features."""
    # Define the formula using the provided dependent and independent variables
    formula = f"{dependent_feature} ~ C(sex) + age + BMI_CALC + {independent_feature}"
    fitted = ols(formula=formula, data=df).fit()
    
    # Create a Series with the results
    result_series = pd.Series({
        "dependent_feature": dependent_feature,
        "independent_feature": independent_feature,
        "beta": fitted.params[independent_feature],
        "t_statistic": fitted.tvalues[independent_feature],
        "p": fitted.pvalues[independent_feature],
        "n": fitted.nobs
    })
    
    return result_series
```

```python
from arivale_data_interface import get_snapshot

# Define your df
df = sleep_merged

# Define your lists of features
dependent_features = sleep_features
independent_features = metabolite_features

# Initialize an empty list to store the results of the regressions
all_results = []

# Iterate through all combinations of dependent and independent features
for dependent_feature in track(dependent_features, description="Dependent Features"):
    for independent_feature in independent_features:
        result = get_results(dependent_feature, independent_feature)
        all_results.append(result)

# concatenate each entry in all_results list and turn into a pandas df
tests = pd.DataFrame(pd.concat(all_results, axis=1))
tests = tests.transpose()

# add the metabolite metadata and calculate corrected p-value, q
anns = get_snapshot("metabolomics_metadata").iloc[:, 0:8]
anns["independent_feature"] = "metabolite_" + anns.CHEMICAL_ID.astype(str)
tests = pd.merge(tests, anns, on="independent_feature")
tests["q"] = multipletests(tests["p"], method="fdr_bh")[1]
```

```python
# save the results to a csv
tests.to_csv('sleep_metabolites_regression_results-sex_age_BMI2.csv', index=False)
```

# Regression w/ sex, age, BMI, and activity covariates

```python
from rich.progress import track
from statsmodels.formula.api import ols
from statsmodels.stats.multitest import multipletests

def get_results(dependent_feature, independent_feature):
    """Get a single association for a specific combination of features."""
    # Define the formula using the provided dependent and independent variables
    activity_covariates = ' + '.join(activity_features)
    formula = f"{dependent_feature} ~ C(sex) + age + BMI_CALC + {activity_covariates} + {independent_feature}"
    fitted = ols(formula, data=df).fit()
    
    # Create a Series with the results
    result_series = pd.Series({
        "dependent_feature": dependent_feature,
        "independent_feature": independent_feature,
        "beta": fitted.params[independent_feature],
        "t_statistic": fitted.tvalues[independent_feature],
        "p": fitted.pvalues[independent_feature],
        "n": fitted.nobs
    })

    return result_series
```

```python
df = sleep_merged

# Define your lists of features
dependent_features = sleep_features
independent_features = metabolite_features

# Initialize an empty list to store the results
all_results = []

# Iterate through all combinations of dependent and independent features
for dependent_feature in track(dependent_features, description="Dependent Features"):
    for independent_feature in independent_features:
        result = get_results(dependent_feature, independent_feature)
        all_results.append(result)

# concatenate each entry in all_results list and turn into a pandas DF
tests = pd.DataFrame(pd.concat(all_results, axis=1))
tests = tests.transpose()

# add the metabolite metadata and calculate corrected p-value, q
anns = get_snapshot("metabolomics_metadata").iloc[:, 0:8]
anns["independent_feature"] = "metabolite_" + anns.CHEMICAL_ID.astype(str)
tests = pd.merge(tests, anns, on="independent_feature")
tests["q"] = multipletests(tests["p"], method="fdr_bh")[1]
```

```python
# save the results to a csv
tests.to_csv('sleep_metabolites_regression_results-activity_sex_age_BMI2.csv', index=False)
```

# Is 2-piperidinone associated with 1,7-dimethylurate?

```python
from rich.progress import track
from statsmodels.formula.api import ols
from statsmodels.stats.multitest import multipletests

def get_results(dependent_feature, independent_feature):
    """Get a single association for a specific combination of features."""
    # Define the formula using the provided dependent and independent variables
    activity_covariates = ' + '.join(activity_features)
    formula = f"{dependent_feature} ~ C(sex) + age + BMI_CALC + {activity_covariates} + {independent_feature} + sleep_efficiency"
    fitted = ols(formula, data=df).fit()
    
    # Create a Series with the results
    result_series = pd.Series({
        "dependent_feature": dependent_feature,
        "independent_feature": independent_feature,
        "beta": fitted.params[independent_feature],
        "t_statistic": fitted.tvalues[independent_feature],
        "p": fitted.pvalues[independent_feature],
        "n": fitted.nobs
    })

    return result_series
```

```python
df = sleep_merged

# Define your lists of features
dependent_features = ['metabolite_100004295'] # 2-piperidinone
independent_features = ['metabolite_100001399'] # 1,7-dimethylurate

# Initialize an empty list to store the results
all_results = []

# Iterate through all combinations of dependent and independent features
for dependent_feature in track(dependent_features, description="Dependent Features"):
    for independent_feature in independent_features:
        result = get_results(dependent_feature, independent_feature)
        all_results.append(result)

# concatenate each entry in all_results list and turn into a pandas DF
tests = pd.DataFrame(pd.concat(all_results, axis=1))
tests = tests.transpose()

# add the metabolite metadata and calculate corrected p-value, q
anns = get_snapshot("metabolomics_metadata").iloc[:, 0:8]
anns["independent_feature"] = "metabolite_" + anns.CHEMICAL_ID.astype(str)
tests = pd.merge(tests, anns, on="independent_feature")
tests["q"] = multipletests(tests["p"], method="fdr_bh")[1]
```

# Diet Assessments, Caffeinated Beverage Intake

```python
diet = get_snapshot('assessments_diet_remapped')
diet.to_csv('diet_assessments.csv', index=False)
caffeine = diet[['public_client_id', 'days_in_program', 'vendor', 'assessment:diet:freq_caffeinated_beverages:int']]
caffeine = caffeine.rename(columns={'vendor': 'vendor_caffeine', 'assessment:diet:freq_caffeinated_beverages:int': 'caffeine_intake'})
caffeine['caffeine_score'] = caffeine['caffeine_intake'].str.extract(r'\((\d+)\)').astype(float)
```

```python
caffeine = caffeine[caffeine.public_client_id.str.isdigit()]
caffeine['public_client_id'] = caffeine.public_client_id.astype('int')
```

```python
# Merge sleep df with measure df from above
sleep_merged = pd.merge_asof(
    caffeine.sort_values(by='days_in_program').dropna(subset='days_in_program'), sleep_merged, 
    by="public_client_id", 
    on="days_in_program", 
    direction="nearest", 
    tolerance=30.0)
```

```python
caffeine_and_derivatives = ['metabolite_849', 'metabolite_100000453', 'metabolite_100001402', 'caffeine_score']
```
